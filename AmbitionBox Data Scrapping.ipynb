```python
import pandas as pd
import requests
from bs4 import BeautifulSoup  #BeautifulSoup is a library which is  used for webscraping in python
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
soup= BeautifulSoup(webpage,'lxml')
```

```python
company = soup.find_all('div',class_='company-content-wrapper')
```

```python
len(company)
```

```python
name=[]
rating=[]
reviews=[]
company_type=[]
headquaters=[]
company_age=[]
employees=[]

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_='rating').text.strip())
    reviews.append(i.find('a',class_='review-count').text.strip())
    company_type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    headquaters.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    company_age.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    
d={'name':name,'rating':rating,'reviews':reviews,'company_type':company_type,'headquaters':headquaters,'company_age':company_age,'employees':employees}
df=pd.DataFrame(d)
    
```

```python
final = pd.DataFrame()
for j in range(1,333):
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(url,headers=headers).text
    
    soup= BeautifulSoup(webpage,'lxml')
    company = soup.find_all('div',class_='company-content-wrapper')
    
    name=[]
    rating=[]
    reviews=[]
    company_type=[]
    headquaters=[]
    company_age=[]
    employees=[]

    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p',class_='rating').text.strip())
        reviews.append(i.find('a',class_='review-count').text.strip())
        try:
            company_type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        except:
            company_type.append(np.nan)
        try:
            headquaters.append(i.find_all('p',class_='infoEntity')[1].text.strip())
        except:
            headquaters.append(np.nan)
        try:
            company_age.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except:
            company_age.append(np.nan)
        try:
            employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)

    d={'name':name,'rating':rating,'reviews':reviews,'company_type':company_type,'headquaters':headquaters,'company_age':company_age,'employees':employees}
    df=pd.DataFrame(d)
    
    final=final.append(df,ignore_index=True)
```

```python
final
```

```python
final.to_csv('Ambition_Box.csv')
```
